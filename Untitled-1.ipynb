```python
import pandas as pd

df = pd.read_excel("mobiliteitsdata.xlsx", engine="openpyxl")
```

```python
df.describe()
```

```python
for kolom in df.select_dtypes(include='number').columns:
    aantal_negatief = (df[kolom] < 0).sum()
    if aantal_negatief > 0:
        print(f"{kolom}: {aantal_negatief} negatieve waarden")
```

```python
new_df = df[
    (df['jaarsalaris_eu'] >= 0) &
    (df['km_auto_per_jaar'] >= 0) &
    (df['km_ov_per_jaar'] >= 0)
]
new_df.shape
```

```python
import numpy as np
try:
    from sklearn.preprocessing import StandardScaler
except ImportError:
    import sys, subprocess
    subprocess.check_call([sys.executable, '-m', 'pip', 'install', 'scikit-learn'])
    from sklearn.preprocessing import StandardScaler

# Maak een kopie zodat de originele waarden behouden blijven
transformed_df = new_df.copy()

# Identificeer numerieke en categorische kolommen
num_cols = transformed_df.select_dtypes(include=['number']).columns.tolist()
cat_cols = transformed_df.select_dtypes(include=['object', 'category']).columns.tolist()

# Log-transformeer sterk scheve numerieke kolommen (absolute skewness > 1)
if num_cols:
    skewness = transformed_df[num_cols].skew().abs()
    skew_cols = skewness[skewness > 1].index.tolist()
else:
    skew_cols = []

for col in skew_cols:
    transformed_df[col + '_log'] = np.log1p(transformed_df[col])
    transformed_df.drop(columns=[col], inplace=True)

# Schaal alle numerieke kenmerken (inclusief nieuw aangemaakte log-kolommen)
num_cols_after = transformed_df.select_dtypes(include=['number']).columns.tolist()
if num_cols_after:
    scaler = StandardScaler()
    transformed_df[num_cols_after] = scaler.fit_transform(transformed_df[num_cols_after])

# One-hot encodeer categorische variabelen (indien aanwezig)
if cat_cols:
    transformed_df = pd.get_dummies(transformed_df, columns=cat_cols, drop_first=True)

# Controle: afmetingen van de getransformeerde dataset
transformed_df.shape

# Toon eerste rijen en kolomnamen
transformed_df.head()
list(transformed_df.columns)

# Sla de getransformeerde dataset op als Excel-bestand
output_path = 'transformed_mobiliteitsdata.xlsx'
transformed_df.to_excel(output_path, index=False, engine='openpyxl')
print('Saved to', output_path)
```

```python
# --- Analyse relaties tussen variabelen ---
try:
    import seaborn as sns
    import matplotlib.pyplot as plt
except ImportError:
    import sys, subprocess
    subprocess.check_call([sys.executable, '-m', 'pip', 'install', 'seaborn', 'matplotlib'])
    import seaborn as sns
    import matplotlib.pyplot as plt
try:
    from statsmodels.stats.outliers_influence import variance_inflation_factor
    from statsmodels.tools.tools import add_constant
except ImportError:
    import sys, subprocess
    subprocess.check_call([sys.executable, '-m', 'pip', 'install', 'statsmodels'])
    from statsmodels.stats.outliers_influence import variance_inflation_factor
    from statsmodels.tools.tools import add_constant

# Numerieke kolommen
num_cols = transformed_df.select_dtypes(include=['number']).columns.tolist()
print('Aantal numerieke features:', len(num_cols))

# Correlatiematrix
corr = transformed_df[num_cols].corr()
corr

# Heatmap
plt.figure(figsize=(12,10))
sns.heatmap(corr, cmap='coolwarm', center=0)
plt.title('Correlation matrix heatmap')
plt.show()

# Top absolute correlaties (excl. diag)
corr_abs = corr.abs()
import numpy as np
np.fill_diagonal(corr_abs.values, 0)
pairs = []
cols = corr_abs.columns
for i in range(len(cols)):
    for j in range(i+1, len(cols)):
        pairs.append((cols[i], cols[j], corr_abs.iloc[i,j]))
pairs_sorted = sorted(pairs, key=lambda x: -x[2])
print('Top 20 sterkste absolute correlaties:')
for a,b,val in pairs_sorted[:20]:
    print(f"{a} <-> {b}: {val:.3f}")

# Pairplot: zorg dat 'co2_uitstoot_per_jaar_KG' altijd inbegrepen is (max 6 features)
plot_cols = []
if 'co2_uitstoot_per_jaar_KG' in num_cols:
    other = [c for c in num_cols if c != 'co2_uitstoot_per_jaar_KG'][:10]
    plot_cols = ['co2_uitstoot_per_jaar_KG'] + other
else:
    plot_cols = num_cols[:6]
sample = transformed_df[plot_cols].sample(n=min(500, len(transformed_df)), random_state=0)
sns.pairplot(sample)
plt.show()

# VIF (controle multicollineariteit)
X = transformed_df[num_cols].dropna()
X_const = add_constant(X)
vif_values = [variance_inflation_factor(X_const.values, i+1) for i in range(len(num_cols))]
vif = pd.Series(vif_values, index=num_cols)
print('Top VIF waarden:')
print(vif.sort_values(ascending=False).head(20))

# Sla correlatiematrix op
corr.to_excel('correlation_matrix.xlsx', engine='openpyxl')
print('Correlation matrix saved to correlation_matrix.xlsx')
```

```python
import numpy as np
try:
    from scipy.stats import norm
except ImportError:
    import sys, subprocess
    subprocess.check_call([sys.executable, "-m", "pip", "install", "scipy"])
    from scipy.stats import norm

co2 = df["co2_uitstoot_per_jaar_KG"].dropna()
mu = co2.mean()
sigma = co2.std(ddof=0)
threshold = 5000
z = (threshold - mu) / sigma
prob = 1 - norm.cdf(z)

print(f"Gemiddelde CO2-uitstoot: {mu:.1f} kg")
print(f"Standaardafwijking: {sigma:.1f} kg")
print(f"Z-score voor 5000 kg: {z:.2f}")
print(f"Kans dat een willekeurig persoon meer dan 5000 kg CO2 uitstoot: {prob:.4f} ({prob*100:.2f}%)")
```
