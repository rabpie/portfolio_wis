```python
import pandas as pd

df = pd.read_excel("mobiliteitsdata.xlsx", engine="openpyxl")
```

```python
new_df = df[
    (df['km_auto_per_jaar'] >= 0) &
    (df['km_ov_per_jaar'] >= 0)
]
new_df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Use the cleaned dataset without negative values
analysis_df = new_df.copy()

# Create figure with subplots for 3 distributions
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Independent variable 1: km_auto_per_jaar (kilometers by car per year)
axes[0].hist(analysis_df['km_auto_per_jaar'], bins=30, color='steelblue', edgecolor='black')
axes[0].set_title('Distribution: km_auto_per_jaar', fontsize=12, fontweight='bold')
axes[0].set_xlabel('km per year')
axes[0].set_ylabel('Frequency')
axes[0].grid(axis='y', alpha=0.3)

# Independent variable 2: km_ov_per_jaar (kilometers by public transport per year)
axes[1].hist(analysis_df['km_ov_per_jaar'], bins=30, color='darkorange', edgecolor='black')
axes[1].set_title('Distribution: km_ov_per_jaar', fontsize=12, fontweight='bold')
axes[1].set_xlabel('km per year')
axes[1].set_ylabel('Frequency')
axes[1].grid(axis='y', alpha=0.3)

# Dependent variable: Find CO2 column (usually co2_per_jaar or similar)
co2_cols = [col for col in analysis_df.columns if 'co2' in col.lower()]
if co2_cols:
    co2_col = co2_cols[0]
    axes[2].hist(analysis_df[co2_col], bins=30, color='green', edgecolor='black')
    axes[2].set_title(f'Distribution: {co2_col}', fontsize=12, fontweight='bold')
    axes[2].set_xlabel('CO2 emissions')
    axes[2].set_ylabel('Frequency')
    axes[2].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

print(f"\nStatistics:")
print(f"km_auto_per_jaar - Mean: {analysis_df['km_auto_per_jaar'].mean():.2f}, Std: {analysis_df['km_auto_per_jaar'].std():.2f}")
print(f"km_ov_per_jaar - Mean: {analysis_df['km_ov_per_jaar'].mean():.2f}, Std: {analysis_df['km_ov_per_jaar'].std():.2f}")
if co2_cols:
    print(f"{co2_col} - Mean: {analysis_df[co2_col].mean():.2f}, Std: {analysis_df[co2_col].std():.2f}")

try:
    from scipy import stats
except ImportError:
    print('\nscipy is not installed; normaalheidstoetsen kunnen niet worden uitgevoerd.')
else:
    samples = {
        'km_auto_per_jaar': analysis_df['km_auto_per_jaar'].dropna(),
        'km_ov_per_jaar': analysis_df['km_ov_per_jaar'].dropna()
    }
    if co2_cols:
        samples[co2_col] = analysis_df[co2_col].dropna()

    print('\nD\'Agostino\'s K^2 normality test:')
    for name, sample in samples.items():
        stat, p = stats.normaltest(sample)
        conclusion = 'niet normaal verdeeld' if p < 0.05 else 'mogelijk normaal verdeeld'
        print(f'{name}: stat={stat:.3f}, p={p:.3f} -> {conclusion}')

    if all(len(sample) <= 5000 for sample in samples.values()):
        print('\nShapiro-Wilk test:')
        for name, sample in samples.items():
            stat, p = stats.shapiro(sample)
            conclusion = 'niet normaal verdeeld' if p < 0.05 else 'mogelijk normaal verdeeld'
            print(f'{name}: stat={stat:.3f}, p={p:.3f} -> {conclusion}')
    else:
        print('\nShapiro-Wilk test wordt overgeslagen voor datasets met meer dan 5000 observaties.')
```
